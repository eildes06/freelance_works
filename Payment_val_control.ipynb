```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data = pd.read_excel('/content/NEXT ALL DATA-Test File.xlsx')
data.info()
```

```python
data["NetReceivedValue"]
```

```python
data["NetReceivedValue"] = data["NetReceivedValue"].str.replace(',', '')
```

```python
data["NetReceivedValue"] = data["NetReceivedValue"].str.replace('.', '',1)
```

```python
data["NetReceivedValue"] = data["NetReceivedValue"].astype(float)
```

```python
data['NetReceivedValue']= data['NetReceivedValue'].fillna("0")
```

```python
a =data['TotalValueTRY'].to_list()
b = data['NetReceivedValue'].to_list()
result = []
```

```python
for y,x in zip(a,b):
    try:
        if x >= y:
          result.append(True)
        else:
          result.append(False)
    except:
          result.append("Bos")
```

```python
data["kiyas"]= result
```
